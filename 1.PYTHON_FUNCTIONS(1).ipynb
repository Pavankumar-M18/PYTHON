# Formal Argument

```python
def add(a,b):       # a & b called formal argument
    c = a+b
    print(c)
add(2,5)           # 5 & 6 called actcual argument
```

# Actual Argument
   - positional argument
   - keyword argument
   - variable length
   - default 

```python
def person(name,age):
    print(name)
    print(age)
person('pavan',24)
```

```python
def person(name,age):
    print(name)
    print(age)
person(22,'pavan')
```

```python
def person(name,age):
    print(name)
    print(age-5)
person(22,'pavan')
```

```python
def person(name,age):
    print(name)
    print(age-5)
person('pavan',24)
```
